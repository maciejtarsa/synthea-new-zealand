```python
import pandas as pd
```

```python
# read the demographics data in
data = pd.read_csv('demographics.csv')
data.head()
```

```python
# extract all areas with their code
timezones = data[['STNAME', 'COUNTY']]
# remove duplicates - only keep 17 areas
timezones = timezones.drop_duplicates()
# rename the columns to match the requirement
timezones = timezones.rename(columns={"STNAME": "STATE", "COUNTY": "ST"})
# add columns for timezone name and code
timezones['TIMEZONE'] = 'New Zealand Standard Time'
timezones['TZ'] = 'NZST'
# print result
timezones
```

```python
# save as csv
timezones.to_csv('timezones.csv', index=False)
```
